```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/noc_regions (1).csv"
noc_regions_df = pd.read_csv(csv_file)
noc_regions_df
```

```python
csv_file = "Resources/athlete_events.csv"
athlete_events_df = pd.read_csv(csv_file)
athlete_events_df
```

```python
#transform athlete dataframe
athlete_cols = ["Name", "Sex", "Age", "Height", "Weight", "Team", "NOC", "Year", "Season", "City", "Sport", "Event", "Medal"]
athlete_transformed = athlete_events_df[athlete_cols].copy()

#rename column headers
athlete_transformed = athlete_transformed.rename(columns= {"Name": 'name',
                                                          "Sex": 'sex',
                                                          "Age": 'age',
                                                          "Height": 'height',
                                                          "Weight": 'weight',
                                                          "Team": 'country',
                                                          "NOC": 'national_olympic_committee',
                                                          "Year": 'year',
                                                          "Season": 'season',
                                                          "City": 'host_city',
                                                          "Sport": 'sport',
                                                          "Event": 'event',
                                                          "Medal": 'medal'})
athlete_transformed
```

```python
csv_file = "Resources/popNIndex.csv"
country_df = pd.read_csv(csv_file)
country_df
```

```python
#transform country dataframe
country_cols= ["Country", "Population", "Consumer_price_index"]
country_transformed = country_df[country_cols].copy()

#rename column headers
country_transformed = country_transformed.rename(columns= {"Country": "country",
                                                          "Population": "population",
                                                          "Consumer_price_index": "cost_of_living"})
country_transformed
```

```python
connection_string = "postgres:postgres@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
#load dataframe into database
athlete_transformed.to_sql(name='athlete', con=engine, if_exists='append', index=False)
```

```python
#load dataframe into database
country_transformed.to_sql(name='country', con=engine, if_exists='append', index=False)
```

```python
#confirm data has been added to postgres
pd.read_sql_query("SELECT * FROM athlete", con=engine).head()
```

```python
#confirm data has been added to postgres
pd.read_sql_query("SELECT * FROM country", con=engine).head()
```

```python
#--1. In descending order, list the frequency count of names
pd.read_sql_query('SELECT name, Count(*) AS frequency FROM athlete GROUP BY name ORDER BY frequency DESC;', con=engine).head()
```

```python
#--4. In descending order list the count of medals "Gold", "Silver", "Bronze", or no medal(null)
pd.read_sql_query('SELECT medal, Count(*) AS frequency FROM athlete GROUP BY medal Order BY frequency DESC;', con=engine).head()
```

```python
#--5. In descending order, list the frequency count of the sex of olympic athletes 
pd.read_sql_query('SELECT sex, Count(*) AS frequency FROM athlete GROUP BY sex ORDER BY frequency DESC;', con=engine).head()
```

```python
#--6. Join the two tables to see comparisons between a country population and if it won a medal
pd.read_sql_query('SELECT * FROM athlete INNER JOIN country ON athlete.country=country.country;', con=engine).head()
```

```python
#--11. In descending order, list the frequency count of the ages of olympic athletes
pd.read_sql_query('SELECT age, COUNT(*) AS frequency FROM athlete GROUP BY age ORDER BY frequency DESC;', con=engine).head()
```
